```python
def image_scrap(driver, loc, category):
    try:
        images = driver.find_element(by=By.XPATH, value='//*[@id="FMP-target"]')
        img_path = []
        for img in images.find_elements(by=By.TAG_NAME,value='img'):
            img_urls = img.get_attribute('src')
            
            id_img = (img_urls.split('/')[-1]).replace(".jpg?im_w=720", "")
            img_name = loc + '_@_' + id_img
    #        img_urls_firstname = '/'.join(img_urls.split('/')[:-3])
    #        img_urls_rename = img_urls_firstname + '/620/466/fitted.jpg'
    #         print(img_urls_rename)
            r = requests.get(img_urls)
            i = Image.open(BytesIO(r.content))
            i.save(f"img/{category}/{img_name}.jpg")
            img_path.append(img_name)
        return img_path
    except:
        print('No images')        
        return 'N/A'
```

```python
def description_scrap(driver):
    try:
        driver.find_element(by=By.XPATH, value='//*[@id="site-content"]/div/div[1]/div[3]/div/div[1]/div/div[4]/div/div[2]/div[2]/button/span/span[1]').click()
        time.sleep(2)
        pop_ele = driver.find_element(by=By.XPATH, value='/html/body/div[11]/section/div/div/div[2]/div/div[2]/div/div/div/div/div/div/div/div/section')
        print(pop_ele.text)
        time.sleep(2)
        cross_ele = driver.find_element(by=By.XPATH, value='/html/body/div[11]/section/div/div/div[2]/div/div[1]/button').click()
    except:
        print("No 'Show more' button")
    
```

```python
def amenities_scrap(driver):
    try:
        amenities_ele = driver.find_element(by=By.XPATH, value='//*[@id="site-content"]/div/div[1]/div[3]/div/div[1]/div/div[6]/div/div[2]/section/div[4]/button').click()
        time.sleep(2)
        amn_ele = driver.find_elements(by=By.CLASS_NAME, value="_11jhslp")
        for i in range(len(amn_ele)):
            print(amn_ele[i].text)
            print('============================')
        time.sleep(2)
        crs_ele = driver.find_element(by=By.XPATH, value='/html/body/div[11]/section/div/div/div[2]/div/div[1]/button').click()
        
    except:
        print("No 'Show all 36 amenities' button")
```

```python
def open_driver():
    op = webdriver.ChromeOptions()
#    op.add_argument('headless')
    s = Service('chromedriver')
    driver = webdriver.Chrome(service=s)
    return driver
```

```python
def vacation_rent_scrap(driver, url, category):
    driver.get(url)
    time.sleep(5)

    # single page data scrapping
    title = title_scrap(driver)
    # if title not found (product sold) don't proceed further for other data
#     if not title:
#         return [None]*7
    
    # scrapping location
    location = loc_scrap(driver)
    
    # scrapping price
    price = price_scrap(driver)
    
    # scrapping rates and reviews
    rate_review = rate_review_scrap(driver)
    
    comments = comment_review_scrap(driver)
    
    # scrapping all images
    images = image_scrap(driver, loc_scrap(driver), category)
    

    # clicking on show more and extracting descriptions
    description = description_scrap(driver)
        
    # clicking on amenities button and extracting information of same
    anemities = amenities_scrap(driver)

    # extracting feature data
    feature = feature_scrap(driver)
    
    return title, location, price, rate_review, comments, images, description, anemities, feature
```

```python
category = 'beach'
with open('img_link.txt', 'r') as f:
    url = f.readlines()
    titles, locations, prices, rates_reviews, comment_lst, img_lst, desc_lst, anm_lst,  feature_lst = [], [], [], [], [], [], [], [], []
    cnt = 1
#     print(single_url)
    for s in url:
        print(cnt, s)
        driver = open_driver()
        
        try:
                title, location, price, rate_review, comments, images, description, anemities, feature = vacation_rent_scrap(driver, s, category)
                if title:
                    titles.append(title)
                    locations.append(location)
                    prices.append(price)
                    rates_reviews.append(rate_review)
                    comment_lst.append(comments)
                    img_lst.append(images)
                    desc_lst.append(description)
                    anm_lst.append(anemities)
                    feature_lst.append(feature)


                    cnt += 1
                    print("=============================================")
#                    clear_output(wait=True)
                    
        except:
                print("no data")
```
